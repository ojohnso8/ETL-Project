```python
# Dependencies
import pandas as pd
```

```python
#take data and read into pandas
cities = "Resources/cities.csv"
data = "Resources/data.csv"

cities_df = pd.read_csv(cities)
data_df = pd.read_csv(data)
```

```python
#display cities_df
cities_df.head()
```

```python
#drop ID column
cities_cleaned_df = cities_df.drop(columns=['id'])
cities_cleaned_df.head()
```

```python
#display data_df
data_df.head()
```

```python
#renamed column in data_df
renamed_data_df = data_df.rename(columns={"U.S. State":"state"})
renamed_data_df.head()
```

```python
#merge on state
merged_data = pd.merge(cities_cleaned_df, renamed_data_df, on="state")
merged_data.head()
```

```python
merged_data.to_csv("merged_data.csv", index=False, header=True)
```
